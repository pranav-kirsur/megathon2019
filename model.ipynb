```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers
```

```python
nFeatures = 10
nDataPoints = 1310
```

```python
dataset = loadtxt("features.csv", delimiter=',') #Sentences x Features

X = dataset[:1100,0:nFeatures]
y = dataset[:1100,nFeatures:]
CV_X = dataset[1100:,0:nFeatures]
CV_Y = dataset[1100:,nFeatures:]
```

```python
model = Sequential()
model.add(Dense(32,kernel_regularizer=regularizers.l2(0.01), input_dim=nFeatures, activation='relu'))
# model.add(Dense(32,   activation='relu'))
# model.add(Dense(20, kernel_regularizer=regularizers.l2(0.01), activation='relu'))
model.add(Dense(20, kernel_regularizer=regularizers.l2(0.01),  activation='relu'))
model.add(Dense(20, kernel_regularizer=regularizers.l2(0.01),  activation='relu'))
model.add(Dense(8, kernel_regularizer=regularizers.l2(0.01),  activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['mean_absolute_error'])
```

```python
model.fit(X, y, epochs=300, batch_size=1100)
```

```python
_, mean_absolute_error = model.evaluate(X, y)
print('Mean Absolute Error Percentage: '+str(mean_absolute_error*100/nDataPoints))
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
import random


def generate_key_list(sentence):
    tokenized = word_tokenize(sentence)
    pos_tagged = nltk.pos_tag(tokenized)
    grammar = "NP: {<JJ|JJS|JJR>*<NN|NNS|NNP|NNPS|CD>}"
    cp = nltk.RegexpParser(grammar)
    noun_phrases_list = [' '.join(leaf[0] for leaf in tree.leaves())
                         for tree in cp.parse(pos_tagged).subtrees()
                         if tree.label() == 'NP']

    return noun_phrases_list


def generate_candidate_keys(noun_phrases):
    possible_keys=[]
    for phrase in noun_phrases:
        tokenized = word_tokenize(phrase)
        pos_tagged = nltk.pos_tag(tokenized)
        print(pos_tagged)
        most_worthy = ""
        for i in pos_tagged:
            if i[1] == 'NN' or i[1] == 'NNS' or i[1]=='NNP' or i[1]=='NNPS':
                most_worthy = i[0]
                break
        for i in pos_tagged:
            if i[1] == 'JJ' or i[1] == 'JJR' or i[1]=='JJS':
                most_worthy = i[0]
                break
        for i in pos_tagged:
            if i[1] == 'CD':
                most_worthy = i[0]
                break
        possible_keys.append(most_worthy)
        if len(possible_keys) == 0:
            possible_keys = random.choice(tokenized)
    return possible_keys

def get_best_key(keys,sentence):
    tokenized = word_tokenize(sentence)
    pos_tagged = nltk.pos_tag(tokenized)
    grammar = "NP: {<JJ|JJS|JJR>*<NN|NNS|NNP|NNPS|CD>}"
    cp = nltk.RegexpParser(grammar)    
    t=cp.parse(pos_tagged)
    n_leaves = len(t.leaves())
    leavepos = set(t.leaf_treeposition(n) for n in range(n_leaves))
    mapping = {}
    for pos in t.treepositions():
        if pos not in leavepos:
            mapping[t[pos].label()] = len(pos)
    pairslist = []
    for i in keys:
        pairslist.append((mapping.get(i,10000),i))
      #  print(pairslist)
    pairslist.sort()
    if len(pairslist) == 0:
        return tokenized[0]
    return pairslist[0][1]



def subs(text,find,replace):
#     find='r'+''""''+find
    output=re.sub(find,replace,text)
    return output

def generate_key_from_sentence(sentence):
    sentence = subs(sentence, '\n', "")
    return get_best_key(generate_candidate_keys(generate_key_list(sentence)),sentence)
```

```python
ab = []
with open('sentences.txt', 'r') as f:
    count = 0
    for lines in f:
        output = {}
        if len(lines) != 0:
            if a[count] == True:
                key = generate_key_from_sentence(lines)
                lines = subs(lines, '\n', " ")
                output['text'] = lines
                output['fib'] = key
                ab.append(output)
        count+=1
```

```python
import json
with open('final.json', 'w') as f:
    for i in ab:
        f.write(json.dumps(i))
        f.write('\n')
```
